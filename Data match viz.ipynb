```python
training = pd.read_table('training_data3.txt')
```

```python
#color code records according to some column, and add a key with labels
sns.lmplot(x='composite_match_score',y='overall_name_score',data=training,hue='is_match',fit_reg=False, scatter_kws={'alpha':0.5},aspect=1.25,size=7)
```

```python
knowland = pd.read_csv('knowland_match_results.csv')
```

```python
#color code records according to some column, and add a key with labels
sns.lmplot(x='Found by matcher',y='Found by fuzzy',data=knowland,hue='Review',fit_reg=False, scatter_kws={'alpha':0.5},aspect=1.25,size=7)
```

```python
knowland_reduced = knowland[(knowland['Found by matcher'] >0) & (knowland['Found by fuzzy'] > 0)]
```

```python
#color code records according to some column, and add a key with labels
sns.lmplot(x='Found by matcher',y='Found by fuzzy',data=knowland_reduced,hue='Review',fit_reg=False, scatter_kws={'alpha':0.5},aspect=1.25,size=7)
```
